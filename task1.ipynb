# 3. Check for Missing Values

```python
data.isna().sum()
```

```python
data = data.dropna() 
```

```python
data.isna().sum()
```

# 4. Remove Duplicates

```python
data.duplicated().sum()
```

```python
data=data.drop_duplicates()
data
```

```python
data.head()
```

# 5. Standardize Text Columns


```python
text_cols = ['STATUS', 'CITY', 'STATE', 'COUNTRY', 'TERRITORY', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME', 'DEALSIZE']

for col in text_cols:
    data[col] = data[col].astype(str).str.lower().str.strip()
```

```python
data['COUNTRY']=data['COUNTRY'].str.capitalize().str.strip()
```

```python
data
```

# 6. Convert Date Formats


```python
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'], errors='coerce')
data
```

# 7. Fix Column Names

```python
data.columns = data.columns.str.lower().str.replace(' ', '_')
data
```

# 8. Ensure Correct Data Types


```python
data['quantityordered'] = pd.to_numeric(data['quantityordered'], errors='coerce')
data['priceeach'] = pd.to_numeric(data['priceeach'], errors='coerce')
data['sales'] = pd.to_numeric(data['sales'], errors='coerce')
data['year_id'] = pd.to_numeric(data['year_id'], errors='coerce')
```

```python
data.info()
```

 # 9. Export Cleaned Data

```python
data.to_csv("cleaned_sales_data.csv", index=False)
```
